```python
import torch
from transformers import pipeline, BitsAndBytesConfig # 导入 BitsAndBytesConfig

# 第二步：加载模型
# 我们使用 unsloth 优化过的 4-bit 量化版 Llama 3，无需 HuggingFace 访问令牌即可直接下载
model_id = "unsloth/llama-3-8b-Instruct-bnb-4bit"

print("正在下载并加载模型，请稍候（约需 2-3 分钟）...")

# 创建推理管道
pipe = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={
        "torch_dtype": torch.float16, # 使用半精度加速
        "quantization_config": BitsAndBytesConfig(load_in_4bit=True), # 将字典替换为 BitsAndBytesConfig 实例
        "low_cpu_mem_usage": True,
    },
)

print("模型加载完成！开始对话...")
```

```python
# 第三步：输入提示词 (Prompt) 并生成
# 这里我们用英文提问（Llama 3 英文能力最强），让它写一段 Python 代码
messages = [
    {"role": "system", "content": "You are a helpful AI assistant for a developer."},
    {"role": "user", "content": "Write a Python function to calculate the Fibonacci sequence using recursion."},
]

prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

# 生成结果
outputs = pipe(
    prompt,
    max_new_tokens=256, # 限制生成长度
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
)

print("-" * 20)
print(outputs[0]["generated_text"][len(prompt):]) # 只打印生成的回答部分
print("-" * 20)
```

```python
import torch
from diffusers import StableDiffusionPipeline

# 第二步：加载绘画模型
print("正在加载 Stable Diffusion 模型...")
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16 # 半精度，节省显存并加速
)
pipe = pipe.to("cuda") # 移动到 GPU 上
```

```python
# 第三步：生成图像
# Prompt: 一只穿着宇航服的猫，赛博朋克风格
prompt = "Generate a programmer sitting at his workstation typing code."
print(f"正在生成: {prompt}")
image = pipe(prompt).images[0]

# 第四步：直接在 Colab 中显示图片
image.show()
# 或者保存图片
image.save("astronaut_cat.png")
```
